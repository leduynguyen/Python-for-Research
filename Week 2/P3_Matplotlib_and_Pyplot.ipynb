### 3.1 Introduction to Matplotlib and Pyplot
Matplotlib is a Python plotting library that produces publication-quality figures. Pyplot is especially useful for interactive work, for example, when you'd like to explore a dataset or visually examine your simulation results. 

Pyplot provides what is sometimes called a state machine interface to matplotlib library. You can loosely think of it as a process where you create figures one at a time, and all commands affect the current figure and the current plot.We will mostly use NumPy arrays for storing the data that we'd like to plot, but we'll occasionally use other types of data objects such as built-in lists.

A basis command is the plt.plot function, which can be used to plot lines and markers. The simplest version of plot has just one argument, and it specifies the y-axis values that are to be plotted. In this case, the plot function, each y-axis value will be plotted against its corresponding index value on the x-axis. Because Python indexing starts at 0, the first element of your array or list appears at location x equal to 0.

```python
# Example 1: Plot with 1 argument
import matplotlib.pyplot as plt

# Plot and list data
plt.plot([1,3,8,11])
plt.show();
```

```python
# Example 2: Plot with 2 arguments
import matplotlib.pyplot as plt
import numpy as np

# Init the data for ploting
x = np.linspace(0, 10, 20)
y = x**2

# Ploting the graph with x and y
plt.plot(x, y)
plt.show();
```

We can also provide a third argument to the plot function, which is a format string that specifies ***color, marker, and line type***. Letters and symbols of the format string are the same as in Matlab.

```python
import matplotlib.pyplot as plt

# Init the data set for ploting
x = np.linspace(0, 10, 20)
y1 = x**2
y2 = x**1.5

# "bo-" -> plt to use blue, circles & solid line.
plt.plot(x, y1, "bo-", linewidth=2, markersize=7)

# "gs-" -> plt to use green, square & solid line.
plt.plot(x, y2, "gs-", linewidth=2, markersize=7)

plt.show()
```

### 3.2 Customizing Your Plots
There are a few important elements that can be easily added to plots:
 * Add a legend with the **legend()**
 * Adjust axes with axis with **axis()**
 * Set axis labels using **xlabel()** and **ylabel()**.
 * Save figure using **savefig()**.

```python
import matplotlib.pyplot as plt
import numpy as np

# Initialize the data set
x = np.linspace(0, 10, 20)
y1 = x**2
y2 = x**1.5

# Plot the data
plt.plot(x,y1,'ro-', linewidth=4, markersize=7, label="First plot")
plt.plot(x,y2,'bs-', linewidth=4, markersize=7, label="Second plot")

# Add label for X and Y axis
plt.xlabel("X value")
plt.ylabel("Y value")

# Scale the axis
plt.axis([-2, 11, -1, 115])

# Add the legend at location upper left
plt.legend(loc="upper left")

# Save the picture to the pdf
plt.savefig("myplot.pdf")

# Show the plot
plt.show()
```

### 3.3 Plotting Using Logarithmic Axes
In some plots, it's helpful to have one or both axes be logarithmic. This means that for any given point to be plotted, its x or y-coordinate, or both, are transformed using the log function.

The plt functions have logarithmic plots are the following:
 * semilogx() plots the x-axes on a log scale and the y in the original scale.
 * semilogy() plots the y-axes on the log scale and the x in the original scale.
 * loglog() plots both x and y on logarithmic scales.

```python
import matplotlib.pyplot as plt
import numpy

x = np.logspace(-1, 1, 20)
y1 = x**2
y2 = x**1.5

# Draw plot using log scale on both X and Y axis
plt.loglog(x, y1, "bo-", linewidth=5, markersize=7, label="First")
plt.loglog(x, y2, "gs-", linewidth=5, markersize=7, label="Second")
plt.legend(loc="lower right")

plt.xlabel("X log")
plt.ylabel("Y log")

plt.show()
```

### 3.4 Generating Histograms
*Example:*

Random normal function to generate 1,000 samples or draws from the standard normal distribution, which is a special case of the normal distribution that has mean equal to 0 and variance equal to 1.

Normalize the histogram and provide locations of bin edges that are used to construct the histogram. By default, hist uses 10 evenly spaced bins and it tries to optimize both bin width and bin locations.

```python
import matplotlib.pyplot as plt
import numpy as np

# Create a random data set
x = np.random.normal(size=1000)

# Plot the histogram
plt.hist(x)

plt.show()
```

We will add first one extra argument, which is **normed to be true**. The histogram, in this case on the y-axis, instead of having the number of observations that fall in each bin, we have the proportion of observations that fall in each bin. That's what it means for a histogram to be normalized. 

To provide the location of the bins, Using a keyword argument called **bins**. Construct the bins using the np.linspace function (21 points = 20 bins between the numbers minus 5 and plus 5.).

```python
import matplotlib.pyplot as plt
import numpy as np

# Create a random data set
x = np.random.normal(size=1000)

# Plot the histogram
plt.hist(x, normed=True, bins=np.linspace(-5, 5, 21))

plt.show()
```

Let's then examine another distribution, which is a bit more exotic, the so-called **gamma distribution**. It is a continuous probability density function that starts at 0 and goes all the way to positive infinity. The gamma distribution, like the normal distribution, has two parameters.

New plt function called **subplot**, which enables us **to have several subplots within each figure**. 
This upload function takes in three arguments:
* The first two specify the **number of rows** and the **number of columns** in the subplot.
* The third argument gives the **plot number**.

For example, if you specify two rows and three columns, then you will have six subplots.

```python
import matplotlib.pyplot as plt
import numpy as np

# Create a data set
x = np.random.gamma(2, 3, 100000)

# Plot the histogram with subplot
plt.figure()

plt.subplot(221)
plt.hist(x, bins=30)
plt.subplot(222)
plt.hist(x, bins=30, normed=True)
plt.subplot(223)
plt.hist(x, bins=30, cumulative=True)
plt.subplot(224)
plt.hist(x, bins=30, cumulative=True, normed=True, histtype="step")

plt.show()
```
